# 2. Исследование зависимостей и формулирование гипотез

### 2.1. Приведем признаки и целевую переменную к численному типу

```python
# Приведем все признаки и целевую переменную к численному типу
df = data.copy()
df['gender'] = df['gender'].replace(['Female', 'Male'], [0, 1])
df['MultipleLines'] = df['MultipleLines'].replace(['No phone service', 'No','Yes'], [0, 1, 2])
df['InternetService'] = df['InternetService'].replace(['No', 'Fiber optic', 'DSL'], [0, 1, 2])
df['Contract'] = df['Contract'].replace(['Month-to-month', 'One year', 'Two year'], [0, 1, 2])
df['PaymentMethod'] = df['PaymentMethod'].replace(['Credit card (automatic)', 'Bank transfer (automatic)',
                                                   'Electronic check', 'Mailed check'], [0, 1, 2, 3])

for feature in ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']:
    df[feature] = df[feature].replace(['No', 'Yes'], [0, 1])
    
for feature in ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies']:
    df[feature] = df[feature].replace(['No internet service', 'No', 'Yes'], [0, 1, 2])
    
df.head()
```

# 3. Построение моделей для прогнозирования оттока

Прежде чем приступать к построению модели, определимся, что необходимо сделать с пропущенными значениями параметра TotalCharges.    

Поскольку данный параметр содержит значения общей суммы, которую клиент заплатил за услуги за все время, наиболее оптимально, если мы заменим пропущенные значения NaN - медианой, по всем объектам выборки. Медиана позволит не учитывать те выбросы, которые могли бы повлиять на среднее значение, если бы мы взяли среднее арифметическое.


```python
# Заменим пропущенные значения параметра TotalCharges
df['TotalCharges'] = df['TotalCharges'].fillna(df['TotalCharges'].median())
```

В основе построения модели используем алгоритм стекинга.  
В качестве базовых алгоритмов используем RandomForestClassifier, GradientBoostingClassifier и LogisticRegression;  
в качестве мета-алгоритма - XGBoost.

```python
# Импортируем библиотеки для построения базовых алгоритмов и мета-алгоритма
from sklearn.model_selection import (GridSearchCV,
                                     train_test_split,
                                     StratifiedKFold)

from sklearn.ensemble import GradientBoostingClassifier, RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from xgboost import XGBClassifier

from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_predict
from sklearn.decomposition import TruncatedSVD
```

#### Базовые алгоритмы

```python
# Сетка параметров для базовых алгоритмов
gbc_params = {'learning_rate': np.arange(0.1, 0.6, 0.1)}
rfc_params = {'n_estimators': range(10, 100, 10),
              'min_samples_leaf': range(1, 5)}
lr_params = {'C': np.arange(0.5, 1, 0.1)}

# Кросс-валидация
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=17)
```

```python
# Разделим выборку
x_train, x_valid, y_train, y_valid = train_test_split(df.drop(columns=['Churn']), 
                                                      df['Churn'],
                                                      train_size=0.7,
                                                      random_state=17)
```

```python
# Создадим объекты базовых моделей
gbc = GradientBoostingClassifier(random_state=17)
rfc = RandomForestClassifier(random_state=17)
lr = LogisticRegression(random_state=17, solver='lbfgs', max_iter=1000)

models = [rfc, lr, gbc]
parameters = [rfc_params, lr_params, gbc_params]
```

---  
Обучим каждую из моделей на тренировочной выборке;  
Определим наиболее оптимальную размерность датафрейма и сочетание параметров для каждого из алгоритмов;  
Построим мета матрицу признаков, состоящую из прогнозов базовых алгоритмов.  

```python
# Отделим признаки и целевую переменную
X = df.drop(columns='Churn')
y = df['Churn']
dimension = [3, 5, 7, 9, 11, 13, 15, 17]
```

```python
# Реализуем функцию, которая поможет подобрать оптимальную размерность для входных данных
# поиск наиболее оптимальных параметров модели
def data_accuracy(model, param, X_transform, y):
    x_train, x_valid, y_train, y_valid = train_test_split(X_transform, y, train_size=0.7, random_state=17)  
    clf = GridSearchCV(estimator=model, param_grid=param, cv=skf)
    clf.fit(x_train, y_train)
    predicted = clf.predict(x_valid)
    best_clf = clf.best_estimator_
    return best_clf, roc_auc_score(predicted, y_valid), predicted

# SVD преобразование размерности входных данных
def dimension_transformation(X, n):
    if n < X.shape[1]:
        svd = TruncatedSVD(n_components=n).fit(X)
        return svd.transform(X)
    else:
        return X
    
# Поиск наиболее оптимальной размерности
def dimension_choice(model, param, X, y, dimension):
    # Возьмем за лучшую размерность исходную размерность датасета
    best_clf, best_accuracy, predict_best_dimension = data_accuracy(model, param, X, y)
    best_dimension = X.shape[1]
    
    for i in dimension:
        X_svd = dimension_transformation(X, i)     
        clf, accuracy, predicted = data_accuracy(model, param, X_svd, y)
        
        if accuracy > best_accuracy:
            best_clf = clf
            predict_best_dimension = predicted
            best_accuracy = accuracy
            best_dimension = i  
    return clf, best_dimension, best_accuracy, predict_best_dimension
```

#### Выбор параметров базовых моделей

```python
%%time

# Определим наиболее оптимальные параметры для базовых моделей и
# определим наиболее оптимальную размерность входных данных для каждой модели
meta_mtrx = np.empty((x_valid.shape[0], len(models)))
best_dimension_models = []
best_models = []

for i, model in enumerate(models):
    clf, dim, accuracy, predicted = dimension_choice(model, parameters[i], X, y, dimension)
    best_models.append(clf)
    best_dimension_models.append(dim)
```

#### Мета-алгоритм

```python
# Сетка параметров для мета-алгоритма
xgb_params = {'n_estimators': range(10, 100, 5),
              'eta': np.arange(0.1, 1., .1),
              'min_child_weight': range(1, 10, 1),
              'subsample': np.arange(0.1, 1., 0.2)}
```

```python
%%time

# Определим наиболее оптимальные параметры для модели мета-алгоритма
xgb = XGBClassifier(random_state=17)
clf_meta = GridSearchCV(xgb, xgb_params, cv=5)
clf_meta.fit(meta_mtrx, y_valid)
best_params = clf_meta.best_params_
```

```python
# Создадим объект модели из самых лучших параметров
xgb = XGBClassifier(random_state=17, eta=best_params['eta'], min_child_weight=best_params['min_child_weight'],
                    n_estimators=best_params['n_estimators'], subsample=best_params['subsample'])
```

#### Стекинг

```python
# Реализуем функцию стекинга
def stacking(best_models, meta_alg, data_train, targets_train, data_test, targets_test=None, random_state=None, test_size=None, cv=5):
    if test_size is None:
        meta_mtrx = np.empty((data_train.shape[0], len(models)))
        
        for n, model in enumerate(best_models):
            svd_train = dimension_transformation(data_train, best_dimension_models[n])
            meta_mtrx[:, n] = cross_val_predict(model, svd_train, targets_train, cv=cv, method='predict')
            model.fit(svd_train, targets_train)
            
        meta_alg.fit(meta_mtrx, targets_train)
        meta_mtrx_test = np.empty((data_test.shape[0], len(models)))
        
        for n, model in enumerate(best_models):
            svd_test = dimension_transformation(data_test, best_dimension_models[n])
            meta_mtrx_test[:, n] = model.predict(svd_test)
            
        meta_predict = meta_alg.predict(meta_mtrx_test)
        return meta_predict
    
    elif test_size > 0 and test_size < 1:
        # Разделим обучающую выборку
        train, valid, train_true, valid_true = train_test_split(data_train, 
                                                                targets_train,
                                                                train_size=test_size,
                                                                random_state=random_state)
        # Построим матрицу из целевых переменных базовых алгоритмов
        meta_mtrx = np.empty((valid.shape[0], len(models)))
        
        for n, model in enumerate(best_models):             
            svd_train = dimension_transformation(train, best_dimension_models[n])
            svd_valid = dimension_transformation(valid, best_dimension_models[n])    
            base_model = model.fit(svd_train, train_true)
            meta_mtrx[:, n] = model.predict(svd_valid)
          
        # Обучим мета-алгоритм
        meta_model = meta_alg.fit(meta_mtrx, valid_true)
        
        # Построим прогноз на тестовых данных
        meta_mtrx_test = np.empty((data_test.shape[0], len(models)))
        
        for n, model in enumerate(best_models):
            svd_data_test = dimension_transformation(data_test, best_dimension_models[n])
            meta_mtrx_test[:, n] = model.predict(svd_data_test)
            
        meta_predict = meta_model.predict(meta_mtrx_test)
        return meta_predict
    else:
        raise ValueError("test_size must be between 0 and 1")
```

# 4. Сравнение качества моделей 

Для сравнения качества моделей используем метрику AUC-ROC

```python
# Качество базовых моделей

for n, model in enumerate(best_models):
    svd_train = dimension_transformation(x_train, best_dimension_models[n])
    svd_valid = dimension_transformation(x_valid, best_dimension_models[n])
    model.fit(svd_train, y_train)
    predicted = model.predict(svd_valid)
    accuracy = roc_auc_score(predicted, y_valid)
    print(model)
    print('Качество модели: %.3f' %accuracy, '\n')
```

```python
# Качество при использовании стекинга
# Построим прогноз
meta_predict  = stacking(best_models, xgb, x_train, y_train, x_valid, targets_test=None, random_state=17, test_size=0.3, cv=5)

# Качество модели
accuracy = roc_auc_score(meta_predict, y_valid)
print(xgb)
print('Качество мета-алгоритма: %.3f' %accuracy)
```

#### Вывод:
Базовые модели показали примерно одинаковый результат.  
При использовании стекинга удалось получить наиболее высокий результат, при этом метаалгоритм обладает большей надежностью.
